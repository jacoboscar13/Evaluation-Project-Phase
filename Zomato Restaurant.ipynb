**1. Target column - Price Range:**

The Price range values are integers and categorical in nature,  ranging from 1 to 4, representing different price categories or ranges.
A lower integer value typically indicates a lower price range (less expensive), while a higher integer value indicates a higher price range (more expensive).
The specific meaning of each integer value in the Price range column may depend on the context of the dataset or the source from which it was obtained.
For example, a common representation might be:

1: Low cost


2: Moderate cost


3: High cost


4: Very high cost

**2. Target column - Average Cost for two:**

The Average Cost for two values are continous and integers, likely representing the local currency (as indicated by the Currency column, which is not included in the provided information).
The values denote the approximate average cost of a meal for two people, which may include charges for food, beverages, and other related expenses.
The specific unit of currency associated with these values is typically mentioned in the Currency column, which is not provided in the excerpt.
For example, if the Currency column indicates the currency is in dollars, and the Average Cost for two is 30, it might mean that the estimated cost for two people is $30.

**Note: Some Outliers might be present in this column**

**3. Aggregate Rating:**

The Aggregate rating is a single numerical value that summarizes the overall rating of a restaurant, combining information from various sources or criteria.
The specific criteria used to calculate the aggregate rating may depend on the context of the dataset or the source from which it was obtained. Common criteria include customer reviews, expert ratings, or a combination of factors like food quality, service, ambiance, etc.
The values in the Aggregate rating column typically range from a minimum to a maximum, and a higher numerical value often indicates a higher overall rating.


For example, if a restaurant has an Aggregate rating of 4.5, it might mean that, based on the specified criteria, the restaurant has received a high overall rating.

**4. Votes:**

The Votes column indicates how many users or customers have rated or voted for a particular restaurant.

A higher value in the Votes column generally suggests that more people have provided feedback or rated the restaurant.
The votes could be related to various aspects of the restaurant, such as overall experience, food quality, service, ambiance, etc.


For example:

If a restaurant has a high number of votes (e.g., 500), it indicates that a substantial number of people have expressed their opinions or rated the restaurant.

If a restaurant has a low number of votes (e.g., 20), it suggests that fewer people have provided feedback.


The Votes column is often used as an indicator of the popularity or engagement level of a restaurant based on customer reviews or ratings.

```python
#creating a function to evaluate models and check their scores
def evaluate_models(models, x_train, x_test, y_train, y_test):
    for name, model in models.items():
        # Fit the model to the training data
        model.fit(x_train, y_train)
        
        # Make predictions on the test data
        y_pred = model.predict(x_test)
        
        # Calculate accuracy
        accuracy = accuracy_score(y_test, y_pred)
        
        # Generate classification report
        class_report = classification_report(y_test, y_pred)
        
        # Generate confusion matrix
        conf_matrix = confusion_matrix(y_test, y_pred)
        
        # Print the results
        print(f"Model: {name}")
        print(f"Accuracy: {accuracy}")
        print("Classification Report:")
        print(class_report)
        print("Confusion Matrix:")
        print(conf_matrix)
        print("-" * 60)
```

```python
#collecting all the models inside by creating a dictionary
models = {"Logistic Regression": LogisticRegression(),"Decision Tree": DecisionTreeClassifier(),"Random Forest Classifier": RandomForestClassifier(),"XG Boost Classifier":xgb.XGBClassifier(),"SVM Classifier": SVC(),"Gradient Boosting Classifier": GradientBoostingClassifier(),"AdaBoost Classifier": AdaBoostClassifier(),"Bagging Classifier": BaggingClassifier(),"ExtraTrees Classifier": ExtraTreesClassifier()}
```

```python
#Checking scores
evaluate_models(models, x_train, x_test, y_train, y_test)
```

# Cross Validation Score

Checking cv score before selecting our final model

```python
#creating a function to check cv score, mean of cv scores, accuracy score and difference between accuracy and cv score to identify the best model

from sklearn.model_selection import cross_val_score

def evaluate_cv(models, x, y, cv=5):
    for name, model in models.items():
        # Calculate cross-validation scores
        cv_scores = cross_val_score(model, x, y, cv=cv)
        
        # Calculate mean of the cross-validation scores
        cv_mean = cv_scores.mean()
        
        # Fit the model to the data
        model.fit(x, y)
        
        # Calculate accuracy score
        accuracy = model.score(x, y)
        
        # Calculate difference between accuracy and cross-validation mean
        diff = accuracy - cv_mean
        
        # Print the results
        print(f"Model: {name}")
        print(f"Cross-validation scores: {cv_scores}")
        print(f"Mean of CV scores: {cv_mean}")
        print(f"Accuracy score: {accuracy}")
        print(f"Difference between accuracy and CV mean: {diff}")
        print("-" * 60)
```

```python
evaluate_cv(models, x, y)
```

we will select Random Forest classifier as our best model as the difference between scores is less

```python
#Creating train and test split at random state = 1
x_train,x_test,y_train,y_test = train_test_split(x1,y1,test_size = 0.30,random_state = maxRS)
```

```python
#importing necessary libraries
from sklearn.ensemble import RandomForestRegressor
from sklearn.tree import DecisionTreeRegressor
from sklearn.neighbors import KNeighborsRegressor as KNN
from sklearn.linear_model import Lasso,Ridge
import xgboost as xgb
from sklearn.svm import SVR
from sklearn.ensemble import ExtraTreesRegressor
```

```python
#collecting all the models inside by creating a dictionary
models = {"Linear Regression": LinearRegression(),"KNN": KNN(),"Random Forest Regressor": RandomForestRegressor(),"XG Boost Regressor":xgb.XGBRegressor(),"Lasso": Lasso(),"Gradient Boosting Regressor": GradientBoostingRegressor(),"Ridge": Ridge(),"Decision Tree Regressor": DecisionTreeRegressor(),"SVR": SVR(),"Extra Trees Regressor":ExtraTreesRegressor()}
```

```python
#creating a function to evaluate models and check their scores
def evaluate_models(models, x_train, x_test, y_train, y_test):
    for name, model in models.items():
        # Fit the model to the training data
        model.fit(x_train, y_train)
        
        # Make predictions on the test data
        y_pred = model.predict(x_test)
        pred_train = model.predict(x_train)
        
        # Calculate R2 score on test data
        accuracy_test = r2_score(y_test,y_pred)*100
        
        # Calculate R2 score on training data
        accuracy_train = r2_score(y_train,pred_train)*100
        
        
        # Generate Mean Absolute Error
        MAE = mean_absolute_error(y_test,y_pred)
        
        # Generate Mean Squared Error
        MSE = mean_squared_error(y_test,y_pred)
        
        # Generate Root Mean Squared Error
        
        RMSE = np.sqrt(mean_squared_error(y_test,y_pred))
        
        # Print the results
        print(f"Model: {name}")
        print(f"R2 score on test data: {accuracy_test}")
        print("R2 score on training data:")
        print(accuracy_train)
        print("Mean Aboslute Error:")
        print(MAE)
        print("Mean Squared Error:")
        print(MSE)
        print("Root Mean Squared Error:")               
        print(RMSE)               
        print("-" * 60)
```

```python
#Checking scores
evaluate_models(models, x_train, x_test, y_train, y_test)
```

# Cross Validation Score¶

Checking cv score before selecting our final model

```python
#creating a function to check cv score, mean of cv scores, accuracy score and difference between accuracy and cv score to identify the best model

from sklearn.model_selection import cross_val_score

def evaluate_cv(models, x1, y1, cv=5):
    for name, model in models.items():
        # Calculate cross-validation scores
        cv_scores = cross_val_score(model, x1, y1, cv=cv)
        
        # Calculate mean of the cross-validation scores
        cv_mean = cv_scores.mean()
        
        # Fit the model to the data
        model.fit(x1, y1)
        
        y_pred =  model.predict(x_test)
        
        # Calculate accuracy score
        R2_score = r2_score(y_test,y_pred)
        
        # Calculate difference between R2 score and cross-validation mean
        diff = R2_score - cv_mean
        
        # Print the results
        print(f"Model: {name}")
        print(f"Cross-validation scores: {cv_scores}")
        print(f"Mean of CV scores: {cv_mean}")
        print(f"R2 score: {R2_score}")
        print(f"Difference between accuracy and CV mean: {diff}")
        print("-" * 60)
```

```python
evaluate_cv(models, x1, y1)
```
