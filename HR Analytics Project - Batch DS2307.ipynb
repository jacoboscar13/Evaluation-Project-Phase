# HR Analytics
Human resource analytics (HR analytics) is an area in the field of analytics that refers to applying analytic processes to the human resource department of an organization in the hope of improving employee performance and therefore getting a better return on investment. HR analytics does not just deal with gathering data on employee efficiency. Instead, it aims to provide insight into each process by gathering data and then using it to make relevant decisions about how to improve these processes.


# Attrition in HR

Attrition in human resources refers to the gradual loss of employees overtime. In general, relatively high attrition is problematic for companies. HR professionals often assume a leadership role in designing company compensation programs, work culture, and motivation systems that help the organization retain top employees.
How does Attrition affect companies? and how does HR Analytics help in analyzing attrition? We will discuss the first question here and for the second question, we will write the code and try to understand the process step by step.


# Attrition affecting Companies

A major problem in high employee attrition is its cost to an organization. Job postings, hiring processes, paperwork, and new hire training are some of the common expenses of losing employees and replacing them. Additionally, regular employee turnover prohibits your organization from increasing its collective knowledge base and experience over time. This is especially concerning if your business is customer-facing, as customers often prefer to interact with familiar people. Errors and issues are more likely if you constantly have new workers.

# T - test

# Years in Current Role and Attrition: 

Employees who have spent more years in their current role are less likely to leave.

Hypothesis:

Null Hypothesis (H0): There is no difference in the ‘YearsInCurrentRole’ between employees who left and those who stayed.

Alternative Hypothesis (H1): There is a difference in the ‘YearsInCurrentRole’ between employees who left and those who stayed.

```python
from scipy import stats

group1 = df[df['Attrition'] == 1]['YearsInCurrentRole']
group2 = df[df['Attrition'] == 0]['YearsInCurrentRole']

# Perform the t-test
t_statistic, p_value = stats.ttest_ind(group1, group2)

# Print the results
print(f'T-statistic: {t_statistic}')
print(f'P-value: {p_value}')
```

The t-statistic value is -6.232038413295282 and the p-value is 6.003185843641377e-10.

The negative t-statistic indicates that the mean ‘YearsInCurrentRole’ for employees who left the company is less than the mean ‘YearsInCurrentRole’ for those who stayed.

The p-value is significantly less than 0.05, which is typically the threshold for statistical significance. This means that we can reject the null hypothesis that there is no difference in the ‘YearsInCurrentRole’ between employees who left and those who stayed.

In other words, the data provides strong evidence to suggest that there is a significant difference in the ‘YearsInCurrentRole’ between employees who left and those who stayed. This could imply that employees who have spent more years in their current role are less likely to leave the company.

# Years Since Last Promotion and Attrition: 

Employees who have not been promoted recently are more likely to leave. 

Hypothesis:

Null Hypothesis (H0): There is no difference in the ‘YearsSinceLastPromotion’ between employees who left and those who stayed.

Alternative Hypothesis (H1): There is a difference in the ‘YearsSinceLastPromotion’ between employees who left and those who stayed.


```python
group_1 = df[df['Attrition'] == 1]['YearsSinceLastPromotion']
group_2 = df[df['Attrition'] == 0]['YearsSinceLastPromotion']

# Perform the t-test
t_statistic, p_value = stats.ttest_ind(group_1, group_2)

# Print the results
print(f'T-statistic: {t_statistic}')
print(f'P-value: {p_value}')
```

The t-statistic value is -1.2657876620135298 and the p-value is 0.2057899591624936.

The negative t-statistic indicates that the mean ‘YearsSinceLastPromotion’ for employees who left the company is less than the mean ‘YearsSinceLastPromotion’ for those who stayed.

However, the p-value is greater than 0.05, which is typically the threshold for statistical significance. This means that we fail to reject the null hypothesis that there is no difference in the ‘YearsSinceLastPromotion’ between employees who left and those who stayed.

In other words, the data does not provide strong evidence to suggest that there is a significant difference in the ‘YearsSinceLastPromotion’ between employees who left and those who stayed. 

This implies that the number of years since the last promotion might not be a strong predictor of attrition, at least not on its own.

```python
wb_attrition = df.groupby(['WorkLifeBalance'])['Attrition'].sum().sort_values(ascending = False)
wb_attrition
```

The null hypothesis and the alternative hypothesis for the t-test based on work-life balance and attrition are as follows:

Null Hypothesis (H0): There is no difference in the ‘WorkLifeBalance’ between employees who left and those who stayed.

Alternative Hypothesis (H1): There is a difference in the ‘WorkLifeBalance’ between employees who left and those who stayed.

```python
group11 = df[df['Attrition'] == 1]['WorkLifeBalance']
group22 = df[df['Attrition'] == 0]['WorkLifeBalance']

# Perform the t-test
t_statistic, p_value = stats.ttest_ind(group11, group22)

# Print the results
print(f'T-statistic: {t_statistic}')
print(f'P-value: {p_value}')
```

The t-statistic value is -2.4548148934297767 and the p-value is 0.014211054989021308.

The negative t-statistic indicates that the mean ‘WorkLifeBalance’ rating for employees who left the company is less than the mean ‘WorkLifeBalance’ rating for those who stayed.

The p-value is less than 0.05, which is typically the threshold for statistical significance. This means that we can reject the null hypothesis and conclude that there is a significant difference in the ‘WorkLifeBalance’ between employees who left and those who stayed.

In other words, the data provides strong evidence to suggest that there is a significant difference in the ‘WorkLifeBalance’ between employees who left and those who stayed.

This could imply that employees who have a better work-life balance are less likely to leave the company.

```python
#finding random state

from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
maxAccu = 0
maxRs = 0
for i in range(1,500):
    x_train,x_test,y_train,y_test = train_test_split(x1,y1,test_size = 0.30,random_state = i)
    RFR = RandomForestClassifier()
    RFR.fit(x_train,y_train)
    pred = RFR.predict(x_test)
    acc = accuracy_score(y_test,pred)
    if acc>maxAccu:
        maxAccu = acc
        maxRS = i 
print('Best Accuracy is =', maxAccu,'at random state =', maxRS )
```
